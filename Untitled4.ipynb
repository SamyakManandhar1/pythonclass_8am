```python
bill = ""
n = int(input("enter n = "))
for i in range(n):
    name= input("enter name ")
    price= int(input("enter price "))
    qty = float(input("enter quantity "))
    total = price * qty
    bill = bill + f'{name}{price}{qty}{total}\n'
print(bill)

file = open ('bill.txt','w')
file.write(bill)
file.close()
```
